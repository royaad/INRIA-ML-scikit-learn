Imagine we are interested in predicting penguins species based on two of their
body measurements: culmen length and culmen depth. First we want to do some
data exploration to get a feel for the data.

What are the features? What is the target?

The data is located in `../datasets/penguins_classification.csv`, load it with
`pandas` into a `DataFrame`.

```python
# Write your code here.
import pandas as pd

df = pd.read_csv('../datasets/penguins_classification.csv')

df.head()
```

Show a few samples of the data.

How many features are numerical? How many features are categorical?

<div class="alert alert-info">
<p class="first admonition-title" style="font-weight: bold;">My Note</p>
<p>
To count the numerical and categorical feature, we can use <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dtypes.html">dtypes</a> attribute. Columns with mixed types (mainly <code>string</code>) are stored with the <code>object</code> dtype. However, this is not always true. Categorical variables can be stored as 0s and 1s. So, it would be better to count <a href="https://pandas.pydata.org/docs/reference/api/pandas.unique.html">unique</a>, as compared to the number of samples.
</p>
</div>

```python
# Write your code here.
# features that are strings 
cat_features = sum(df.dtypes == 'object')
num_features = sum(df.dtypes != 'object')
print(f"We have {num_features} numerical features, and {cat_features} categorical features.")
```

Show features distribution for each class. Hint: use
[`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

```python
# Write your code here.
import seaborn as sns

columns = ["age", "education-num", "hours-per-week"]
_ = sns.pairplot(
    data=df,
    vars=df.columns[:-1],
    hue=df.columns[-1],
    plot_kws={"alpha": 0.2},
    height=3,
    diag_kind="hist",
    diag_kws={"bins": 30},
)
```
